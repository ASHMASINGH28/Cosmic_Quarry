```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
img=cv2.imread("C4.PNG",cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap="gray")
plt.title("carbon spectrum")
plt.axis("off")
plt.show()
```

```python
denoised = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(denoised,cmap="gray")
plt.axis("off")
plt.show()
```

```python
intensity = np.mean(denoised, axis=0)
pixels = np.arange(len(intensity))
```

```python
pixel_known = [100, 700]
wavelength_known = [400, 700]
```

```python
wavelengths = np.interp(pixels, pixel_known, wavelength_known)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(wavelengths, intensity, color='black')
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity")
plt.title("Spectrograph: Intensity vs Wavelength")
plt.grid(True)
plt.show()
```
