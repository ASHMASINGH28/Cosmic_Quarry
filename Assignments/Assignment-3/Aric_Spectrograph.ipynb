#### Image Processing

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
img=cv2.imread(r"C:\Users\SUNNY DAVID\Pictures\Download\Au.png")                #Denoising the image
kernel = np.array([[-1,-1,-1], [-1, 9, -1], [-1, -1, -1]])                      #Kernel for sharpening filter
sharpened_img = cv2.filter2D(img, -1, kernel)    
dn_img=cv2.fastNlMeansDenoisingColored(sharpened_img, None, 10, 10, 7, 25) 
cv2.imwrite(r"C:\Users\SUNNY DAVID\Pictures\Download\Au_sharpened.png",dn_img)      
cv2.imshow("window",sharpened_img)
cv2.waitKey(0)
```

#### Generating Spectrograph of Gold

```python
image = cv2.imread(r"C:\Users\SUNNY DAVID\Pictures\Download\Au_sharpened.png")
rows=image.shape[0] 
columns=image.shape[1]
midrow = image.shape[0] // 2                # Middle row of image
spectrum_line =image[midrow,:,:]            # Taking the middle row of image for consideration
intensity=[]
for i in range(columns):
    R=spectrum_line[i,2]
    G=spectrum_line[i,1]
    B=spectrum_line[i,0]
    x=0.2126*R + 0.7152*G + 0.0722*B        # Calculating Intensity
    intensity.append(x)
wavelengths=np.linspace(400,701,columns)    # Making wavelength coordinates
plt.figure(figsize=(10, 5))
plt.plot(wavelengths, intensity, color='black')
plt.title("Spectrograph of Gold")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity")
plt.grid(True)
plt.savefig(r"C:\Users\SUNNY DAVID\Pictures\Download\Specg_Au.png")
plt.show()
```
