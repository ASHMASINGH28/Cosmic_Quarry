#Image preprocessing




```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("225.jpeg")
height, width, channels = img.shape

def img_preprocessing(img):
  #denoising using gaussian blur
  denoised = cv2.GaussianBlur(img, (5, 5), 0)

  #sharpening image
  box_filter = cv2.blur(denoised, (5,5))
  sharpened = cv2.subtract(denoised, box_filter)
  processed_img = cv2.add(denoised , sharpened)

  return processed_img

img = img_preprocessing(img)

#displaying image
plt.imshow(img)
plt.title('Processed image')
plt.show()
```

#Obtaining spectrograph from a spectral image

```python
#assuming one pixel column as single wavelength
#defining each pixel value as intensity
def calc_intensity(R, G, B):
  return 0.299*R+0.587*G+0.114*B

def spectrograph(img):
  height, width, channels = img.shape
  intensity = []
  for i in range(width):
    column = calc_intensity(img[:, i, 0], img[:, i, 1], img[:, i, 2])
    intensity.append(np.sum(column))

  wavelength = []
  for i in range(width):
    wavelength.append(400 + i*(300/width))  #callibrating axes

  return intensity, wavelength

intensity, wavelength = spectrograph(img)

#plotting qualitatively
plt.plot(wavelength, intensity)
plt.xlabel("Wavelength")
plt.ylabel("Intensity")
plt.title("Spectrograph")
plt.grid(True)
plt.show()
```

#Expanding to a dataset

```python
#Atomic spectrum of H

img = cv2.imread("hydrogen.png")
img = img_preprocessing(img)                #image preprocessing
intensity, wavelength = spectrograph(img)   #obtaing intensities and corresponding wavelengths

#plotting qualitatively
plt.plot(wavelength, intensity)
plt.xlabel("Wavelength")
plt.ylabel("Intensity")
plt.title("Spectrograph of H")
plt.grid(True)
plt.show()
```
